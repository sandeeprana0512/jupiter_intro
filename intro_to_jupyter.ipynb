## this is a python code to calculate Factorial of a given number

```python
def factorial(x):
    fact = 1
    while(x > 1):
        fact *= x
        x = x-1
    return fact
print(factorial(5))
```

```python
print(factorial(4))
```

Different file formats of Jupyter Notebook serve different purposes and may be useful in various scenarios:

.ipynb (IPython Notebook):

This is the default file format for Jupyter Notebook. Useful for creating interactive documents containing live code, equations, visualizations, and explanatory text. Ideal for sharing notebooks with colleagues, collaborators, or posting them on platforms like GitHub. .html (HTML):

Exporting a notebook to HTML format can be useful for creating static, self-contained versions of the notebook. Useful for sharing notebooks with individuals who may not have Jupyter Notebook installed, as HTML files can be viewed in any web browser. .pdf (PDF):

Exporting a notebook to PDF format is helpful when creating printable, formatted documents. Useful for sharing finalized reports, presentations, or academic papers with a wider audience, as PDF files maintain consistent formatting across different platforms. .py (Python script):

Converting a notebook to a Python script (.py) is beneficial when you want to distribute the code without the narrative or visualizations. Useful for software development workflows, version control systems, or automation tasks where a pure Python script is required. .md (Markdown):

Exporting a notebook to Markdown format can be useful for integrating code and narrative into documentation, blogs, or websites. Markdown files are lightweight, easy to read, and can be easily converted to various formats using markdown processors. Each file format has its own advantages and use cases, and the choice depends on the intended audience, purpose of sharing, and the level of interactivity or formatting required in the document.

## Using widgets (creating a input slider widget and displaying its plot)

```python
import ipywidgets as widgets
from IPython.display import display
import numpy as np
from bokeh.plotting import figure, output_notebook, show
from bokeh.models import ColumnDataSource
from bokeh.layouts import column

# Initialize Bokeh output
output_notebook()

# Create a slider widget
number_slider = widgets.IntSlider(
    value=50,
    min=1,
    max=100,
    description='Select a number:'
)

# Create an output widget to display the plot
output = widgets.Output()

# Define a function to update the plot based on the selected number
def update_plot(change):
    with output:
        output.clear_output(wait=True)
        selected_number = change['new']  # Access the 'new' value from the change dictionary

        # Create a simple plot
        x = np.linspace(0, 10, 100)
        y = np.sin(selected_number * x)

        # Create a ColumnDataSource for Bokeh
        source = ColumnDataSource(data=dict(x=x, y=y))

        # Create a Bokeh plot
        p = figure(title=f'Plot for selected number: {selected_number}', x_axis_label='X-axis', y_axis_label='Y-axis')
        p.line('x', 'y', source=source, line_width=2)

        # Display the plot
        show(p)

# Attach the update_plot function to the slider's value change event
number_slider.observe(update_plot, names='value')

# Display the widgets
display(number_slider)
display(output)

# Initialize the plot
update_plot({'new': number_slider.value})




```
